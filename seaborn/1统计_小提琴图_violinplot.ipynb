seaborn.violinplot() 是一个用于绘制小提琴图( violin plot )的函数，它结合了箱线图和核密度估计的特点，可以展示数据的分布形状以及重要的统计量。

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

海冰范围数据

```python
sns.set_theme()
seaice = pd.read_csv("seaborn-data/seaice.csv")
display(seaice.sample(10))
seaice["Date"] = pd.to_datetime(seaice["Date"])
seaice["Decade"] = seaice["Date"].dt.year.round(-1)
sns.violinplot(
    data=seaice,    # 指定数据源
    x="Extent",     # x轴表示海冰范围（数值变量）
    y="Decade",     # y轴表示年代（分类变量）
    orient="y",     # 指定小提琴的方向为垂直（默认是x轴为主方向）,小提琴将在y轴（Decade）的每个类别上水平展开
    hue="Decade",   # 添加hue参数
    palette="Blues",# 示例调色板
    fill=True,      # 小提琴内部不填充颜色（只显示轮廓）
    legend=False
    )

plt.title("Sea Ice Extent Distribution by Decade")
plt.xlabel("Ice Extent (million sq km)")
plt.ylabel("Decade")
```

这段代码使用了Seaborn库创建了一个嵌套且分裂的小提琴图，用于比较不同组别之间的数据分布。

tips.csv是Seaborn自带的一个示例数据集，包含餐厅小费数据。典型列包括：total_bill(总账单), tip(小费), sex(性别), smoker(是否吸烟), day(星期几), time(用餐时间), size(用餐人数)

```python
sns.set_theme(style="dark")
tips = pd.read_csv("seaborn-data/tips.csv")
display(tips.sample(10))

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(
    data=tips,           # 使用tips数据集
    x="day",             # x轴为星期几
    y="total_bill",      # y轴为账单金额
    hue="smoker",        # 按是否吸烟分组
    palette={"Yes": "g", "No": ".35"},  # 自定义颜色
    # 分裂显示组别，将同一分类下的两个hue组别(吸烟/不吸烟)显示在同一个小提琴的左右两侧
    # 如果不设置split=True，两组会分开显示为两个独立的小提琴
    split=True,
    inner="quart",       # 在小提琴内部显示四分位线（25%、50%、75%分位数）
    fill=False,          # 不填充颜色
)
```

这段代码使用Seaborn和Pandas对脑网络相关性数据进行处理并绘制小提琴图。

```python
sns.set_theme(style="whitegrid")

# Load the example dataset of brain network correlations
df = pd.read_csv("seaborn-data/brain_networks.csv", header=[0, 1, 2], index_col=0)
display(df.sample(10))

# Pull out a specific subset of networks
used_networks = [1, 3, 4, 5, 6, 7, 8, 11, 12, 13, 16, 17]
used_columns = (df.columns.get_level_values("network")
                          .astype(int)
                          .isin(used_networks))
df = df.loc[:, used_columns]

# Compute the correlation matrix and average over networks
corr_df = df.corr().groupby(level="network").mean()
corr_df.index = corr_df.index.astype(int)
corr_df = corr_df.sort_index().T

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 6))

# Draw a violinplot with a narrower bandwidth than the default
sns.violinplot(data=corr_df, bw_adjust=.5, cut=1, linewidth=1, palette="Set3")

# Finalize the figure
ax.set(ylim=(-.7, 1.05))
sns.despine(left=True, bottom=True)
```
